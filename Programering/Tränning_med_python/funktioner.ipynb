```python
def medelvärde(a,b,c,d):
    medel = (a+b+c+d)/4 #lokal variabel, syns ej utanför funtionen
    return medel

print(medelvärde(1,2,3,4))
```

```python
 # räkna ord
 def räkna_ord(text):
     ord = text.split()
     antal_ord = len(ord)
     return antal_ord

citat = "I stand on the shoulders of giants"
print(f"Antal ord i citatet: {räkna_ord(citat)}")
```

```python
def minsta(tal1, tal2):
    if tal1 > tal2:
        return tal2
    else:
        return tal1

print(minsta(1,1000000))
```

# Discord uppgift

Idag finns det 7 miljarder människor på jorden. En modell som beskriver antal människor på jorden som funktionen av tiden

```python
import numpy as np

def N(t):
    n = 11 / (1 + (3.4 * np.exp(-0.03 * t)))
    return n

print(N(0))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def N(t):
    n = 11 / (1 + (3.4 * np.exp(-0.03 * t)))
    return n

år = 1000

t = np.arange(år)
plt.xlabel("År efter 1950")
plt.ylabel("Antal människor i miljarder")
plt.title("Modell för population")
plt.plot(t,N(t))
plt.grid()
plt.show()

print(f"Övre gräns: {N(1000)} miljarder")
```

```python
n = int(input("bestäm n där sista talet blir 2n + 1 "))

def Summa(n):
    s = 0
    for i in range(1,2*n+2,2):
        s += i
    return s

print(f"Summan är {Summa(n)}")
```
