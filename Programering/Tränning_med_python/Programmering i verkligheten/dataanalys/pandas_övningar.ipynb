```python
#1 a)

import pandas as pd
import numpy as np


df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)
förvärvsarbetande = df[df["sysselsättning"] == "förvärvsarbetande"]
Antal_förvärvsarbetande = förvärvsarbetande["2019"].sum(axis=0)

print(Antal_förvärvsarbetande)
```

```python
#1 b)

ej_förvärvsarbetande = df[df["sysselsättning"] == "ej förvärvsarbetande"]
Antal_ej_förvärvsarbetande = ej_förvärvsarbetande["2019"].sum(axis=0)

print(Antal_ej_förvärvsarbetande)
```

```python
#1 c)

import matplotlib.pyplot as plt

labels = "ej_förvärvsarbetande", "förvärvsarbetande"
size = [Antal_ej_förvärvsarbetande, Antal_förvärvsarbetande]

fig1, ax1 = plt.subplots()
ax1.pie(size, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal') 

plt.show()
```

```python
#2 a)

förvärvsarbetande = df[df["sysselsättning"] == "förvärvsarbetande"]
män_förvärvsarbetande = förvärvsarbetande[förvärvsarbetande["kön"] == "män"]

Antal_män_förvärvsarbetande = män_förvärvsarbetande["2019"].sum(axis=0)
print(Antal_män_förvärvsarbetande)
```

```python
#2 b)

förvärvsarbetande = df[df["sysselsättning"] == "förvärvsarbetande"]
kvinnor_förvärvsarbetande = förvärvsarbetande[förvärvsarbetande["kön"] == "kvinnor"]

Antal_kvinnor_förvärvsarbetande = kvinnor_förvärvsarbetande["2019"].sum(axis=0)
print(Antal_kvinnor_förvärvsarbetande)
```

```python
#2 c)

ej_förvärvsarbetande = df[df["sysselsättning"] == "ej förvärvsarbetande"]
män_ej_förvärvsarbetande = ej_förvärvsarbetande[ej_förvärvsarbetande["kön"] == "män"]

Antal_män_ej_förvärvsarbetande = män_ej_förvärvsarbetande["2019"].sum(axis=0)
print(Antal_män_ej_förvärvsarbetande)
```

```python
#2 d)

ej_förvärvsarbetande = df[df["sysselsättning"] == "ej förvärvsarbetande"]
kvinnor_ej_förvärvsarbetande = ej_förvärvsarbetande[ej_förvärvsarbetande["kön"] == "kvinnor"]

Antal_kvinnor_ej_förvärvsarbetande = kvinnor_ej_förvärvsarbetande["2019"].sum(axis=0)
print(Antal_kvinnor_ej_förvärvsarbetande)
```

```python
#2 e)

labels = "män ej förvärvsarbetande", "män förvärvsarbetande", "kvinnor ej förvärvsarbetande", "kvinnor förvärvsarbetande"
size = [Antal_män_ej_förvärvsarbetande, Antal_män_förvärvsarbetande, Antal_kvinnor_ej_förvärvsarbetande, Antal_kvinnor_förvärvsarbetande]

plt.bar(labels, size)
plt.show()
```
