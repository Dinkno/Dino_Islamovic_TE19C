# Miljöutsläpp

Ett miljöutsläpp halverar fågelpopulation i en ö varje år.

Fågelpopulation började på 8000st

Hur många år tar det tills populationen blir mindre än en tiondel?

```python
fåglar = 8000
antal_år = 0


while fåglar >= 800:
    fåglar /= 2
    antal_år += 1
    print(f"År {antal_år}: antal fåglar: {fåglar}st")

print(f"Det tar {antal_år} år tills fågelpopulationen understiger en tiodel")
```

```python
'''
1L mjölk 1500000 bakterier
I rumstemp ökar a talet bakterier med 50% varje timme
Mjölken blir sur när bakterier överstiger 10000000 st
Hur många timmar tills mjölken blir sur
'''
import matplotlib.pyplot as plt
import numpy as np

bakterier = 1.5e6 # 1.5*10**6
surt = 1e7
faktor = 1.5
timmar = 0

y_bakterier = [bakterier] # lista som ska hålla bakterier "förändring"


while bakterier < surt:
    bakterier = faktor*bakterier
    y_bakterier.append(bakterier) # lägger på bkterier till vår lista
    timmar += 1

x_timmar = np.arange(timmar+1)

print(f"Det tar {timmar}h i rumstemperatur för att mjölken ska surna")
print(f"x = {x_timmar}")
print(f"y = {y_bakterier}")

plt.plot(x_timmar, y_bakterier, "*-")
plt.xlabel("Timmar (h)")
plt.ylabel("Antal bakterier")
plt.title("bakterietillväxten i mjölken i rumstemperatur")
plt.show()
```
